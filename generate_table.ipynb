```python
import tensorflow as tf
import pandas as pd
from tensorflow.contrib.predictor import from_saved_model
import s3fs
```

```python
fs = s3fs.S3FileSystem()
```

```python
def search_image(root_path, all_files = []):
    files = fs.ls(root_path) 
    for file in files:
        if not fs.isdir(file):
            all_files.append(file)
        else:
            search_image(file,all_files)
    return all_files
```

```python
subdirs = fs.ls('s3:...')
absroot = []
for d in subdirs:
    if (d[-10:-4] == 'xdfp11'):
        absroot.append(d)
for i in absroot:
    files_11 = search_image(i)
```

```python
files_11[:10]
```

```python
result_list = []
for file in files_11:
    sub_list=[]
    Protein = 'xdfp'
    sub_list.append(Protein)
    Tray = file[28:38]
    sub_list.append(Tray)
    Date = file[39:45]
    sub_list.append(Date)
    Observation = 'None'
    sub_list.append(Observation)
    Well = file[57:60]
    sub_list.append(Well)
    Well_Position = file[61]
    sub_list.append(Well_Position)
    P = file[63:len(file)+1]
    P_1 = P.find('_')
    method = P[:P_1]
    sub_list.append(method)
    A= P.rfind('_')
    Time = P[A-6:A]
    sub_list.append(Time)
    After_time = P[A+1]
    sub_list.append(After_time)
    Full_path = file
    sub_list.append(Full_path)
    result_list.append(sub_list) 
result_list
```

```python
files_11 = pd.DataFrame(result_list, columns=['Protein','Tray','Date','Observation','Well','Well Position','method','Time','After time','Full path'])
```
